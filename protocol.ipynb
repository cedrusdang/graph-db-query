```python
def transform_time_to_hour(data, col_name = 'time'):
    # Left only hour instead of hh:mm format
    data['hour'] = data[col_name].str.split(":").str[0].astype(int)
    # Drop the old time column
    data = data.drop(columns=["time"])
    return data
data = transform_time_to_hour(data)
data
```

```python
def create_table(data: pd.DataFrame, cols_sets: dict, table_name:str, sk:bool = True) -> pd.DataFrame:
    cols = cols_sets[table_name]
    # Extract columns from the data frame and reset the index
    df_table = data[cols].drop_duplicates().sort_values(cols[0]).reset_index(drop=True)
    # sk enable pipeline
    if sk:
        # Create sk keys
        df_table = df_table.reset_index(drop=False).rename(columns={"index": f"{table_name}_sk"})
        df_table.iloc[:, 0] = df_table.iloc[:, 0] + 1
        # Merge dim table to data table to add sk
        data = data.merge(df_table, on=cols, how='left')
        # Drop the orginal cols
        data = data.drop(columns=cols)
    return df_table, data

# Dims and columns  sets
dim_cols_sets = {
    "date": ["dayweek", "day_of_week", "month", "year", ],
    "crash_type": ["crash_type"],
    "holiday": ["easter", "christmas"],
    "lga": ["lga"],
    "sa4": ["sa4"],
    "nra": ["nra"],
    "state": ["state"],
    "gender": ["gender"],
    "age_group": ["age_group"],
    "road_user": ["road_user"],
    "involvement": ["bus", "heavy_rigid_truck", "articulated_truck"],
    "road_type": ["road_type"],
    "speed_limit": ["speed_limit"],
    "time": ["hour"]
}

# Create dimension tables
dim_date, data = create_table(data, dim_cols_sets, "date")
dim_crash_type, data = create_table(data, dim_cols_sets, "crash_type")
dim_holiday, data = create_table(data, dim_cols_sets, "holiday")
dim_lga, data = create_table(data, dim_cols_sets, "lga")
dim_sa4, data = create_table(data, dim_cols_sets, "sa4")
dim_nra, data = create_table(data, dim_cols_sets, "nra")
dim_state, data = create_table(data, dim_cols_sets, "state")
dim_gender, data = create_table(data, dim_cols_sets, "gender")
dim_age_group, data = create_table(data, dim_cols_sets, "age_group")
dim_road_user, data = create_table(data, dim_cols_sets, "road_user")
dim_involvement, data = create_table(data, dim_cols_sets, "involvement")
dim_road_type, data = create_table(data, dim_cols_sets, "road_type")
dim_speed_limit, data = create_table(data, dim_cols_sets, "speed_limit", sk=False)
dim_time, data = create_table(data, dim_cols_sets, "time", sk=False)

# Fact and columns  sets
dim_fact_sets = {
    "crash": ["crash_id", "Number Fatalities"],
    "fatality": ["fatality_sk"],
}

# Create Fact tables
fact_crash, data = create_table(data, dim_fact_sets, "crash")
fact_fatality, data = create_table(data, dim_fact_sets, "fatality", sk=False)
```
